```python
# now we will check the comparison of price and phone brand
plt.figure(figsize=(20,10))
sns.barplot(x='brand_name', y='price', data=df_smartphone)
plt.xticks(rotation=90)
plt.title('Comparison of Price and Phone Brand')
plt.show()
```

```python
# Unique values for each categorical column
cat_cols = df_smartphone.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"{col}: {df_smartphone[col].nunique()} unique values")
    print(df_smartphone[col].unique()[:10])  # Show first 10 unique values
    print()
```

```python
# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(df_smartphone['price'], kde=True)
plt.title('Price Distribution')
plt.xlabel('Price (INR)')
plt.show()
```

```python
# User ratings distribution
plt.figure(figsize=(8,5))
sns.histplot(df_smartphone['rating'], bins=20, kde=True)
plt.title('User Ratings Distribution')
plt.xlabel('User Rating')
plt.show()
```

```python
# Price vs RAM
plt.figure(figsize=(8,5))
sns.scatterplot(x='ram_capacity', y='price', data=df_smartphone)
plt.title('Price vs RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (INR)')
plt.show()
```

```python
# Comparison of Processor Speed vs Processor Brand
plt.figure(figsize=(12,6))
sns.boxplot(x='processor_brand', y='processor_speed', data=df_smartphone)
plt.xticks(rotation=45)
plt.title('Processor Speed vs Processor Brand')
plt.xlabel('Processor Brand')
plt.ylabel('Processor Speed (GHz)')
plt.show()
```

```python
# 3D scatter plot: Screen Size vs Refresh Rate vs Battery Capacity

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    df_smartphone['screen_size'], 
    df_smartphone['refresh_rate'], 
    df_smartphone['battery_capacity'],
    c=df_smartphone['battery_capacity'], cmap='viridis', alpha=0.7
)
ax.set_xlabel('Screen Size (inches)')
ax.set_ylabel('Refresh Rate (Hz)')
ax.set_zlabel('Battery Capacity (mAh)')
ax.set_title('Screen Size vs Refresh Rate vs Battery Capacity')
plt.show()
```

 ***Conclusion from the Smartphone Dataset     (Based on EDA)*** <br>
---

 1. Mid-Range is King
The sweet spot for value lies in the 15,00025,000 range.

These phones consistently offer:

>5G support

>5064 MP rear cameras

>Fast charging (~33W+)

>68GB RAM with 128GB storage

>120Hz displays

And they dont tank in user ratingsoften score 80+.

 TL;DR: You dont need to burn 50k+ for solid performance anymore.

 2. High Specs  High Ratings
Ratings are not strictly tied to price or specs.

Some budget phones score higher than flagships, likely due to user satisfaction, value-for-money, and fewer bugs.

 TL;DR: A 19k Realme might beat a 40k Samsung in day-to-day happiness.

 3. 5G is the New Normal
Nearly every phone, even under 15k, supports 5G now.

Its no longer a premium featureits expected.

 TL;DR: If you're buying a 4G-only phone in 2025... youre basically future-proofing for the past.

 4. Battery Size is Flattening
5000mAh is now industry standard.

Real differentiation comes from fast charging (some up to 120W!).

No strong correlation between higher battery and better ratings.

 TL;DR: Battery size isnt everythingcharging speed and software optimization matter more.

 5. Snapdragon Still Reigns Supreme
Snapdragon processors dominate both performance and presence.

Dimensity and Exynos are competitive, but Snapdragon models trend toward higher user ratings.

 TL;DR: Youre generally safe picking Snapdragon unless youre on a tight budget.

 6. 120Hz Display is the New Standard
Most phones now ship with 90Hz or 120Hz refresh rates.

60Hz is now basically outdated.

 TL;DR: If youre buying a phone with 60Hz in 2025, you're doing your eyes dirty.

 7. Camera Specs Are Marketing Bait
Tons of phones claim 108MP sensorsbut real image quality depends on processing.

50MP64MP cameras with solid software often outperform higher-MP sensors.

 TL;DR: Dont fall for megapixel flexingcheck real-world performance and reviews.

 8. 128GB + 6GB RAM is the Norm
Most devices now come with 128GB internal storage, which is becoming the new baseline.

Expandable storage is missing in almost half the devicesespecially in higher-end ones.

 TL;DR: If you want expandable storage, dont go too premium. Weird, but true.

 9. Feature Trade-Offs Still Exist
Budget phones skip on NFC, IP ratings, stereo speakers, etc.

Even some mid-range models ditch IR blasters and microSD slots.

 TL;DR: You still need to prioritize. Can't get everything in a 15k phone (yet).

 10. Trends Are Leveling Off
The wild jumps in specs (20182022) have slowed.

Now it's more about incremental improvements:

Charging speed

Software optimization

Camera processing

Thermals & build quality

 TL;DR: Unless your current phone sucks, upgrading every year isn't worth it anymore.

 ***Final Thought:*** <br>
This dataset shows we're in a spec maturity phase now. The market isnt about just raw numbers anymoreits about experience per rupee. And the mid-range phones? Theyve seriously stepped up.

You dont have to buy premium to feel premium anymore.
